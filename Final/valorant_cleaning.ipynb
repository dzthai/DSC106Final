```python
import re

import requests
from bs4 import BeautifulSoup
```

```python
region = {
    "na": "north-america",
    "eu": "europe",
    "ap": "asia-pacific",
    "la": "latin-america",
    "la-s": "la-s",
    "la-n": "la-n",
    "oce": "oceania",
    "kr": "korea",
    "mn": "mena",
    "gc": "game-changers",
    "br": "Brazil",
    "ch": "china",
}

headers = {
    "User-Agent": "Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:52.0) Gecko/20100101 Firefox/52.0",
}
```

```python
def vlr_team(team_num):
        URL = "https://www.vlr.gg/team/" + str(team_num)
        html = requests.get(URL, headers=headers)
        soup = BeautifulSoup(html.content, "lxml")
        status = html.status_code

        base = soup.find(class_=re.compile("col-container"))
        containers = base.find_all(class_="team-header")

        result = []
        for container in containers:
            # column 1
            name = container.find(class_=re.compile("team-header-name")).text.strip()

            # column 2 - logo
            logo_container = container.find(class_=re.compile("wf-avatar team-header-logo"))
            img = logo_container.find("img")
            logo = "https:" + img["src"]

            # column 2 - ctry
            country = container.find(class_=re.compile("team-header-country")).text.strip()

            result.append(
                {
                    "name": name,
                    "country": country,
                    "logo": logo,
                }
            )

        data = {"data": result}

        if status != 200:
            return "nope"
        return data
```

```python
import sys 

teams1 = list(set(df['Team1ID']))
teams2 = df['Team2ID']
```

```python
result = {}

for i in range(len(teams1)):
    data = vlr_team(teams1[i])
    if (data == 'nope'):
        continue
    doot = pd.json_normalize(data)  
    if ('\n' in doot['data'][0][0]['name']):
        abbrev = doot['data'][0][0]['name'].split('\n')[1]
        if abbrev == '(inactive)':
            continue
        name = doot['data'][0][0]['name'].split('\n')[0]
        sys.stdout.write(abbrev + " " + name + '\n')
        result[abbrev] = name
    else:
        result[doot['data'][0][0]['name']] = doot['data'][0][0]['name']
```

```python
abbreviations= df['TeamAbbreviation']
teamNames = []
for i in abbreviations:
    if (i not in result.keys()):
        teamNames.append(i)
    else:
        teamNames.append(result[i])
```
